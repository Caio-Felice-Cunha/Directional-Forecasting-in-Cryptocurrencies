### 2.1.1 Log Transformation of Price Features:
The price features (open, high, low, close) have skewed distributions, so applying a log transformation will help to normalize the data, especially for models that assume normality in features. A log transformation compresses the range and reduces the effect of outliers by transforming large values into smaller ones.

np.log1p is used for log transformation to handle zeros (since log(0) is undefined, but log1p(0) is valid and returns 0).

This should help to make the distributions more normal, reducing the impact of extreme values (outliers).

```python
# Apply log transformation to the price features
train_df['log_open'] = np.log1p(train_df['open'])
train_df['log_high'] = np.log1p(train_df['high'])
train_df['log_low'] = np.log1p(train_df['low'])
train_df['log_close'] = np.log1p(train_df['close'])

# Plot the transformed distributions
train_df[['log_open', 'log_high', 'log_low', 'log_close']].hist(bins=50, layout=(2, 2), figsize=(12, 8))
plt.suptitle("Log-Transformed Distribution of OHLC Features")
plt.show()
```

### 2.1.2 Handling Outliers:
Outliers, particularly in financial data, can sometimes distort the model's predictions. I wll either:

First: Cap extreme values using the quantiles (e.g., cap at the 99th percentile):

```python
# Capping at the 99th percentile for the OHLC features
upper_bound = 0.99
train_df['open'] = np.clip(train_df['open'], train_df['open'].quantile(0.01), train_df['open'].quantile(upper_bound))
train_df['high'] = np.clip(train_df['high'], train_df['high'].quantile(0.01), train_df['high'].quantile(upper_bound))
train_df['low'] = np.clip(train_df['low'], train_df['low'].quantile(0.01), train_df['low'].quantile(upper_bound))
train_df['close'] = np.clip(train_df['close'], train_df['close'].quantile(0.01), train_df['close'].quantile(upper_bound))
```

Second: Remove extreme outliers using the IQR (Interquartile Range) method:

Little note: Clipping ensures that extreme values are replaced with the boundary values, avoiding the influence of large outliers.

IQR-based removal eliminates outliers that fall outside 1.5 times the interquartile range (a common technique for outlier removal).

```python
# Removing outliers based on IQR
Q1 = train_df[['open', 'high', 'low', 'close']].quantile(0.25)
Q3 = train_df[['open', 'high', 'low', 'close']].quantile(0.75)
IQR = Q3 - Q1

train_df = train_df[~((train_df[['open', 'high', 'low', 'close']] < (Q1 - 1.5 * IQR)) |
                      (train_df[['open', 'high', 'low', 'close']] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Histogram for price features
plt.figure(figsize=(12, 6))
train_df[['open', 'high', 'low', 'close']].hist(bins=50, layout=(2, 2), figsize=(12, 8))
plt.suptitle("Capped, Without outliers and Log-Transformed Distribution of OHLC Features")
plt.show()

# KDE for volume and number of trades
plt.figure(figsize=(10, 5))
sns.kdeplot(train_df['volume'], label='Volume', fill=True) #shade=True)
sns.kdeplot(train_df['number_of_trades'], label='Number of Trades', fill=True) #shade=True)
plt.title('Distribution of Volume and Number of Trades')
plt.legend()
plt.show()
```

### 2.1.3 Handling Zero or Low Trading Volume/Number of Trades:
Since there are many periods with zero or very low trading volumes, I want to create features or flags to handle these periods. Here are a few ideas:

* Flag zero volume/trades: Create binary features to indicate when the volume or number of trades is zero.
* Remove periods with no activity: If periods of no trading activity are not relevant to my prediction task, I can consider removing them.

Here's how I created flags for zero or low volume:

The binary flag features helps mt model distinguish between periods of normal and abnormal activity.
This approach can help the model better understand how price movement correlates with trading activity.

```python
# Create binary flags for periods with zero volume or number of trades
train_df['zero_volume_flag'] = (train_df['volume'] == 0).astype(int)
train_df['low_trades_flag'] = (train_df['number_of_trades'] == 0).astype(int)

# Create flags for low trading activity (e.g., below a threshold)
threshold_volume = train_df['volume'].quantile(0.01)
threshold_trades = train_df['number_of_trades'].quantile(0.01)

train_df['low_volume_flag'] = (train_df['volume'] < threshold_volume).astype(int)
train_df['low_trades_flag'] = (train_df['number_of_trades'] < threshold_trades).astype(int)
```

### 3.3 Resample Data for Aggregation
Depending on the granularity of your data, you may want to resample it to a higher level, such as daily, weekly, or monthly, to see broader trends.

For example, if the data is minute-level, you can aggregate it to daily:

```python
# Convert 'timestamp' to datetime
train_df['timestamp'] = pd.to_datetime(train_df['timestamp'])

# Set 'timestamp' as the index
train_df.set_index('timestamp', inplace=True)
```

```python
# Resample to daily frequency by taking the mean for price features and summing the volume
daily_data = train_df.resample('D').agg({
    'open': 'mean',
    'high': 'mean',
    'low': 'mean',
    'close': 'mean',
    'volume': 'sum'
})

# Display the first few rows
print(daily_data.head())
```

```python
# Ensure there are no missing values
daily_data['close'] = daily_data['close'].ffill()

# Apply decomposition on the 'close' price data
decomposition = seasonal_decompose(daily_data['close'], model='multiplicative', period=30)  # period is approximate based on frequency

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot the decomposed time series
plt.figure(figsize=(10, 8))
plt.subplot(411)
plt.plot(daily_data['close'], label='Original')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='upper left')
plt.tight_layout()

# Save the figure
plt.savefig('C:/Users/caiof/Downloads/decomposed.png')

plt.show()

```
